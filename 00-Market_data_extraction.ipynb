```python
start = dt.datetime(2021,7,1)
end = dt.datetime(2022,7,31)
```

```python
tickers = ['BTC-USD'] 
```

```python
# for ticker in tickers:
#     data = web.DataReader(ticker, 'yahoo', start, end)
#     data.to_csv('{}.csv'.format(ticker))
```

```python
data = []
```

```python
for index, ticker in enumerate(tickers):
    data.append(yf.download(tickers=ticker,start=start, end=end, interval="60m"))
    data[index].to_csv('{}.csv'.format(ticker))
```

```python
for index, ticker in enumerate(tickers):
    fig_dims = (20, 10)
    fig, ax = plt.subplots(figsize=fig_dims)
    p = sns.lineplot(x = 'Datetime', y = 'Close', data = data[index], ax = ax)
    p.set_title('{} Close Price'.format(ticker))
```
