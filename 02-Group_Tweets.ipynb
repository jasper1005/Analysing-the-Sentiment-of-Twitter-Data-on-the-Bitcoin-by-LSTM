### Read all extracted tweets

```python
import os
import pandas as pd

# Directory where the files are stored
dir_path = 'Data-Input/02-Extracted_tweets/'

# Get a list of all files in the directory
files = os.listdir(dir_path)

# Filter the list for .csv files
csv_files = [file for file in files if file.endswith('.csv')]

# Sort the list of files (if files are named in a way that they sort in chronological order)
csv_files.sort()

# Initialize an empty list to store dataframes
frames = []

# Loop through the list of .csv files
for file in csv_files:
    # Read the .csv file and append it to the list
    frames.append(pd.read_csv(dir_path + file, delimiter="\t"))

# Concatenate all dataframes in the list
tweets = pd.concat(frames)
```

```python
# tweets = tweets.drop(['Unnamed: 0','User','Tweet ID', 'Source'],axis=1)
tweets = tweets.drop(['Unnamed: 0', 'Source'],axis=1)
tweets = tweets.reset_index(drop=True)
```

```python
tweets.drop(tweets[(tweets['Lang'] != 'en')].index, inplace=True)
tweets = tweets.reset_index(drop=True)
```

### Remove URL, Tag, Number, all content started as RT.  

```python
def remove_url(txt):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",txt).split())
    #  Returns the same txt string with url's removed.
```

```python
tweets['Tweets'] = [remove_url(tweet) for tweet in tweets['Tweets']]
```

```python
tweets['Tweets'] = [" ".join(re.sub("RT", "", tweet).split()) for tweet in tweets['Tweets']]
tweets['Tweets'] = [" ".join(re.sub("[0-9]", "", tweet).split()) for tweet in tweets['Tweets']]
```

```python
tweets = tweets.reset_index(drop=True)
```

### Extract the missed content by User Name (Number of Friends and Followers)
##### Non-exist user will be counted as 0 friend and 0 follower

```python
import tweepy
import pandas as pd
import csv
import re 
import string
 
consumer_key = 'consumer_key'
consumer_secret = 'consumer_secret'
access_key = 'access_key'
access_secret = 'access_secret'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key,access_secret)

api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
#tweets.loc[index, 'Normalize_Like'] = (tweet['Like'] - tweets['Like'].mean()) / tweets['Like'].std()
#tweets.loc[index, 'Normalize_Retweet'] = (tweet['Retweet'] - tweets['Retweet'].mean()) / tweets['Retweet'].std()

# the screen_name of the targeted user
list_followers = []
list_friends = []
for index, tweet in tweets.iterrows(): 
    try:
        user = api.get_user(screen_name = tweet['User'])
        followers_count = user.followers_count
        friends_count = user.friends_count
        print(str(index) + ': ' + str(followers_count) + ', ' + str(friends_count))
    
    except tweepy.errors.NotFound:
        list_followers.append(1)
        list_friends.append(1)
        continue
    except tweepy.errors.Forbidden:
        list_followers.append(1)
        list_friends.append(1)
        continue
    except tweepy.errors.TweepError:
        list_followers.append(1)
        list_friends.append(1)
        continue
    else:# if no error
        if followers_count == 0:
            list_followers.append(1)
        else:
            list_followers.append(followers_count)
            
        if friends_count == 0:
            list_friends.append(1)
        else:
            list_friends.append(friends_count)
        continue
```

```python
tweets['Friends'] = list_friends
tweets['Followers'] = list_followers
```
