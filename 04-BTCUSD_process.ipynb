### Change Time format from timestamp to Date and Time

```python
btcusd = pd.read_csv('Data-Input/04-btcusd.csv', delimiter = ',')
```

```python
btcusd
```

```python
btcusd['Dates'] = pd.to_datetime(btcusd['time'], unit='ms').dt.date
btcusd['Time'] = pd.to_datetime(btcusd['time'], unit='ms').dt.time
```

```python
btcusd
```

### Change Timestamp to Hour:00 format for matching the seniment data

Since Sentiment data is collected by hour. Therefore, the sentiment data will copy to every minute within an hour

```python
column_names = ["Dates", "Time", "open", "close", "high", "low", "volume", "time"]
```

```python
btcusd = btcusd.reindex(columns=column_names)
```

```python
btcusd
```

```python
btcusd['time'] = pd.to_datetime(btcusd['time'], unit='ms')
```

```python
#btcusd.style.format({"time": lambda t: t.strftime("%H:00")})
```

```python
btcusd["time"] = btcusd["time"].dt.strftime('%H:00')
```
