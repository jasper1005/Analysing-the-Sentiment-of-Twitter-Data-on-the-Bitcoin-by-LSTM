# Extracting Tweets

The search_30_day() function from Twitter API had been applied to extract the data within a month.

tweets_as_df() function selected the necessary data from the original data which collected by Twitter API.

```python
search_words ='BTC'
numTweets = 100

def extract(handle,since,end):
    # premium search
    tweets = tweepy.Cursor(api.search_30_day, label='textproject2' , 
                           query=handle, fromDate=since, toDate=end).items(numTweets)
    print('tweets downloaded from {}'.format(handle), 'start: ', since, ' ,end: ', end)
    return tweets
```

```python
def tweets_as_df(handle,since,end):
    columns = set() 
    allowed_types = [str, int]
    n = 0
    tweets_data = []
    for status in extract(handle,since,end):
        if status.in_reply_to_status_id is not None:
            is_reply = True
        else:
            is_reply = False
        status_dict = dict(vars(status))
        keys = status_dict.keys()
        single_tweet_data = {"User": status.user.screen_name,
                             "Length": len(status.text), # extracting length of the text in tweet
                             "Date": status.created_at.strftime("%Y-%m-%d"),
                             "Time": status.created_at.strftime("%H:%S"),
                             "isRetweet": hasattr(status, 'retweeted_status'), # checking retweet or not 
                             "isReply": is_reply,
                             "image_URL": status.entities} # checking reply or not
        for k in keys:
            try:
                v_type = type(status_dict[k]) 
            except:
                v_type = None
            if v_type != None:
                if v_type in allowed_types: # loop all information from user timeline into single_tweet_data dictionary
                    single_tweet_data[k] = status_dict[k]
                    columns.add(k) 
        tweets_data.append(single_tweet_data) # add it into list as data
    header_cols = list(columns) 
    header_cols.append("User")
    header_cols.append("Length")
    header_cols.append("Date")
    header_cols.append('isRetweet')
    header_cols.append('isReply')
    header_cols.append('Time')
    df = pd.DataFrame(tweets_data, columns=header_cols)
    return df
```

### Download tweets by hour

```python
import datetime

# Initialize an empty DataFrame to store tweets
tweets = pd.DataFrame()

# Start and end date
start_date = datetime.datetime.strptime('202208200000', '%Y%m%d%H%M')
end_date = datetime.datetime.strptime('202208210000', '%Y%m%d%H%M')

# Increment of 1 hour
delta = datetime.timedelta(hours=1)

while start_date < end_date:
    # Format start and end times as strings
    start_str = start_date.strftime('%Y%m%d%H%M')
    end_str = (start_date + delta).strftime('%Y%m%d%H%M')
    
    # Append tweets to DataFrame
    tweets = tweets.append(tweets_as_df(search_words, start_str, end_str))
    
    # Advance start_date by 1 hour
    start_date += delta
```
