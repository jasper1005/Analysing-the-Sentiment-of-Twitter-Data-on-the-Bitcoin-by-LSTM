```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

### Read Grouped Tweets

```python
tweets = pd.read_csv('Data-Input/group_tweets.csv', delimiter = "\t")
tweets = tweets.drop(['Unnamed: 0'],axis=1)
```

### Convert function for add space between word

```python
def convert(s): 
  
    # initialization of string to "" 
    new = "" 
  
    # traverse in the string 
    for x in s: 
        new += x + ' ' 
  
    # return string 
    return new 
```

### Stopword cleanning

```python
# example_sent = "This is a sample sentence, showing off the stop words filtration."
# Stop word clear
stop_words = set(stopwords.words('english'))

for index, tweet in tweets.iterrows(): 
    result=[]
    word_tokens = word_tokenize(str(tweet['Tweets']))
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    filtered_sentence = []
  
    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)

    tweets.loc[index, 'Tweets'] = convert(filtered_sentence)
    print(tweets.loc[index, 'Tweets'])
```

### Extract 'compound' score from VADER analyzer 

```python
tweets.insert(1,'Sentiment',0)
```

```python
for index, tweet in tweets.iterrows(): 
    vs = analyzer.polarity_scores(str(tweet['Tweets']))
    tweets.loc[index, 'Sentiment'] = vs['compound']
```

### Normalize and Weight the sentiment score

##### Sentiment = Sentiment * (1 + Normalize_LikeCount) * (1 + Normalize_RetweetCount) * (1 + Normalize_Weight_followers[followers/friends])

```python
from sklearn import preprocessing
import numpy as np
```

```python
def normalize(df,f1,f2):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[f1].max()
        min_value = df[f1].min()
        result[f2] = (df[f1] - min_value) / (max_value - min_value)
    return result[f2]
```

```python
tweets['Normalize_Like'] = normalize(tweets, 'Like', 'Normalize_Like')
tweets['Normalize_Retweet'] = normalize(tweets, 'Retweet', 'Normalize_Retweet')
```

```python
tweets['Weight_followers'] = 0
```

```python
weight = 0
for index, tweet in tweets.iterrows(): 
    tweets.loc[index, 'Weight_followers'] = tweet['Followers'] / tweet['Friends']
```

```python
tweets['Normalize_Weight_followers'] = normalize(tweets, 'Weight_followers', 'Normalize_Weight_followers')
```

```python
print(max(tweets['Normalize_Weight_followers']))
print(max(tweets['Normalize_Like']))
print(max(tweets['Normalize_Retweet']))
```

```python
tweets['Final_Sentiment'] = tweets['Sentiment'] * (1 + tweets['Normalize_Like'] * tweets['Normalize_Retweet'] * tweets['Normalize_Weight_followers'])
```
